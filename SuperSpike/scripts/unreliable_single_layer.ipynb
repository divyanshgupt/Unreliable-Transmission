```python
import numpy as np
import torch
from matplotlib import pyplot as plt
import pickle
import pandas as pd
```

### Load Loss Recordings

#### Spk_freq = 20 Hz

#### Probability = 0.9

Learning Rate = 1e-3, 10e-3, 50e-3

```python
loss_9_1e3 = "Organized data/Spk_freq=20/2022-03-31 08 rate = 0.001 p=0.9/loss_rec epsilon= 0.0001learning_rate = 0.001spike freq = 20"
loss_rec_9_1e3 = pd.read_pickle(loss_9_1e3)

loss_9_10e3 = "Organized data/Spk_freq=20/2022-03-31 07 rate = 0.001 p=0.9/loss_rec epsilon= 0.0001learning_rate = 0.001spike freq = 20"
loss_rec_9_10e3 = pd.read_pickle(loss_9_10e3)

loss_9_50e3 = "Organized data/Spk_freq=20/2022-03-25 19 rate = 0.05 p=0.9/loss_rec epsilon= 0.0001learning_rate = 0.05spike freq = 20"
loss_rec_9_50e3 = pd.read_pickle(loss_9_50e3)

loss_9_100e3 = "Organized data/Spk_freq=20/2022-03-29 21 rate = 0.1 p=0.9/loss_rec epsilon= 0.0001learning_rate = 0.1spike freq = 20"
loss_rec_9_100e3 = pd.read_pickle(loss_9_100e3)

fig, axs = plt.subplots(4, 1, sharex=True, figsize=(12,15))

for i in range(10):
    axs[0].plot(loss_rec_9_1e3[i], alpha=0.6)
axs[0].plot(loss_rec_9_1e3.mean(axis=0), color='black', label='Avg loss')
# axs[0].set_xlabel("Epoch No.")
axs[0].set_ylabel("Loss")
axs[0].set_title("20Hz, p=0.9, R_0 = 1e-3")

for i in range(10):
    axs[1].plot(loss_rec_9_10e3[i], alpha=0.6)
axs[1].plot(loss_rec_9_10e3.mean(axis=0), color='black', label='Avg loss')
# axs[1].set_xlabel("Epoch No.")
axs[1].set_ylabel("Loss")
axs[1].set_title("20Hz, p=0.9, R_0 = 10e-3")

for i in range(10):
    axs[2].plot(loss_rec_9_50e3[i], alpha=0.6)
axs[2].plot(loss_rec_9_50e3.mean(axis=0), color='black', label='Avg loss')
axs[2].set_xlabel("Epoch No.")
axs[2].set_ylabel("Loss")
axs[2].set_title("20Hz, p=0.9, R_0 = 50e-3")
axs[2].set_ylim(0, 250)
for i in range(10):
    axs[3].plot(loss_rec_9_100e3[i], alpha=0.6)
axs[3].plot(loss_rec_9_100e3.mean(axis=0), color='black', label='Avg loss')
axs[3].set_xlabel("Epoch No.")
axs[3].set_ylabel("Loss")
axs[3].set_title("20Hz, p=0.9, R_0 = 100e-3")
axs[3].set_ylim(0, 250)
```
